```python
df = pd.read_sql('''
  select 
    sd.id,
    sd.peso_l,
    sd.peso_ch,
    sd.peso_cn,
    sd.peso_m,
    sd.peso_r,
    sd.nota_minima_l,
    sd.nota_minima_ch,
    sd.nota_minima_cn,
    sd.nota_minima_m,
    sd.nota_minima_r,
    sd.media_minima,
    sd.nota_l,
    sd.nota_ch,
    sd.nota_cn,
    sd.nota_m,
    sd.nota_r,
    sd.nota_l_com_peso,
    sd.nota_ch_com_peso,
    sd.nota_cn_com_peso,
    sd.nota_m_com_peso,
    sd.nota_r_com_peso,
    sd.nota_candidato,
    sd.nota_corte,
    st.*
  from sisu_nota sd
  tablesample bernoulli(10)
  inner join sisu_tempo st on sd.dimensao_tempo = st.id
  ''', con=CONN)
```

```python
df_ano = df[['nota_l',
    'nota_ch',
    'nota_cn',
    'nota_m',
    'nota_r',
    'nota_candidato',
    'ano'
]]


df_ano = df_ano.melt(id_vars='ano', value_vars=['nota_l',
    'nota_ch',
    'nota_cn',
    'nota_m',
    'nota_r',
    'nota_candidato'
], var_name='notas', value_name='pontuacao')
plt.figure(figsize=(14, 6))
sns.boxplot(x='ano', y='pontuacao', hue='notas', data=df_ano, dodge=True)
```
